```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the file paths
item_info_path = r'..\dataset\item_info.csv'
transactions_info_path = r'..\dataset\transactions_info.csv'

# Load the data
item_info_df = pd.read_csv(item_info_path)
transactions_info_df = pd.read_csv(transactions_info_path)

# Data Visualization

# Visualize the distribution of item quantities sold
plt.figure(figsize=(10, 5))
sns.histplot(transactions_info_df['item_qty'], bins=20, kde=True)
plt.title('Distribution of Item Quantities Sold')
plt.xlabel('Quantity Sold')
plt.ylabel('Frequency')
plt.show()

# Visualize relationships between features
plt.figure(figsize=(10, 5))
sns.scatterplot(x='item_qty', y='invoice_time', data=transactions_info_df)
plt.title('Scatter Plot: Quantity Sold vs. Invoice Time')
plt.xlabel('Quantity Sold')
plt.ylabel('Invoice Time')
plt.show()

# Visualize item categories
plt.figure(figsize=(10, 5))
sns.countplot(x='item_category', data=item_info_df)
plt.title('Item Categories Count')
plt.xticks(rotation=90)
plt.xlabel('Item Category')
plt.ylabel('Count')
plt.show()
```
